```python
# 단위 분리 및 숫자로 변환 ,데이터 타입 float64
# Mileage	Engine	Power
df['Mileage'] = df['Mileage'].str.replace(" km/kg", "", regex=False).str.replace(" kmpl", "", regex=False).astype(float)
df['Engine'] = df['Engine'].str.replace(" CC", "").astype(float)
df['Power'] = df['Power'].str.replace(" bhp", "").astype(float)
```

### 히트맵과,회귀 적합선을 보니 'Engine','Power'가 양의 상관관계를 가진다.

```python
# Price와 나머지 변수들만 선택하여 새로운 데이터프레임 생성
df_selected = df[['Price', 'Year', 'Kilometers_Driven', 'Engine','Power', 'Seats', 'Mileaget']]

# 상관계수 행렬 계산
correlation_matrix = df_selected.corr()

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 회귀 적합선 추가 : kind= "reg"
sns.pairplot(df,y_vars="Price",x_vars=['Year', 'Kilometers_Driven', 'Engine','Power','Seats', 'Mileaget'],kind="reg")
plt.ylim(-10000,290000)
```

```python
# 학습용 / 평가용 데이터 분리
df_train,df_test = train_test_split(df,test_size=0.3,random_state=77)

print("학습용 데이터 크기 : {}".format(df_train.shape))
print("평가용 데이터 크기 : {}".format(df_test.shape))
# 선형 회귀 분석
reg_model = smf.ols(formula = "Price ~  Year+ Kilometers_Driven+ Engine+Power+ Seats+ Mileaget",data=df_train)

# 적합 
reg_result = reg_model.fit()
print(reg_result.summary())
```

```python
# 잔차 분석
import statsmodels.api as sm

# 잔차 정규성 검정 (QQ plot)
sm.qqplot(reg_result.resid, line='s')
plt.show()

# 잔차의 히스토그램
plt.hist(reg_result.resid, bins=30)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# 다중공선성 평가 (VIF 계산)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 독립 변수들만 선택
X = df_train[['Year', 'Kilometers_Driven', 'Engine', 'Power', 'Seats', 'Mileaget']]
X = sm.add_constant(X)

# VIF 계산
vif = pd.DataFrame()
vif['Variable'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
import statsmodels.api as sm

# 독립 변수와 종속 변수 설정
X_train = df_train['Price']
y_train = df_train['Engine']

# 상수 추가
X_train = sm.add_constant(X_train)

# 회귀 모델 적합
reg_model = sm.OLS(y_train, X_train).fit()

# 회귀 모델 결과 출력
print(reg_model.summary())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 예측값 계산
X_test = df_test['Price']
X_test_with_const = sm.add_constant(X_test)
y_pred = reg_model.predict(X_test_with_const)

# 회귀선 시각화
plt.figure(figsize=(10, 6))
plt.scatter(df_test['Price'], df_test['Engine'], label='Actual Data')
plt.plot(df_test['Price'], y_pred, color='red', label='Fitted Line')
plt.xlabel('Price')
plt.ylabel('Engine')
plt.title('Linear Regression: Engine vs Price')
plt.legend()
plt.show()
```

```python
# 중위수 계산
price_median = df_test['Price'].median()
power_median = df_test['Power'].median()

# 산점도 그래프
plt.figure(figsize=(10, 6))
plt.scatter(df_test['Price'], df_test['Power'], label='Actual Data')
plt.axvline(x=price_median, color='r', linestyle='--', label='Price Median')
plt.axhline(y=power_median, color='b', linestyle='--', label='Power Median')
plt.xlabel('Price')
plt.ylabel('Power')
plt.title('Scatter Plot: Price vs Power with Medians')
plt.legend()
plt.show()
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Name 열을 범주형으로 변환
df['Name'] = df['Name'].astype('category')

# ANOVA 모델 적합
anova_model = ols('Price ~ Name', data=df).fit()
anova_results = anova_lm(anova_model)

# ANOVA 결과 출력
print(anova_results)
```

```python
# ANOVA 가설 검정
model = ols('Price ~ C(First_Word)', data=df).fit()  # C()를 사용하여 범주형 변수로 처리
anova_table = sm.stats.anova_lm(model, typ=2)  # ANOVA 테이블 생성
anova_table
```

```python
# 브랜드별 평균 가격 계산
average_prices = df.groupby('First_Word')['Price'].mean().sort_values()

# 브랜드를 가격이 낮은 순서로 정렬
df['First_Word'] = pd.Categorical(df['First_Word'], categories=average_prices.index, ordered=True)
df_sorted = df.sort_values('First_Word')

# 시각화 (가격이 낮은 브랜드별)
plt.figure(figsize=(12, 8))
sns.boxplot(x='First_Word', y='Price', data=df_sorted, order=average_prices.index)
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Price Distribution by Brand (Low to High)')
plt.show()
```
